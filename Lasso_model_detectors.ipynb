```python
import csv
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt 
import matplotlib.font_manager
from sklearn import svm
from sklearn.covariance import EllipticEnvelope 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import LocalOutlierFactor
from sklearn import linear_model as lm
```

```python
from sklearn.datasets import load_diabetes
```

Import danych.

```python
data = load_diabetes()
```

```python
data
```

```python
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target
```

```python
X.info()
```

Model lasso

```python
model_lasso = lm.Lasso() #(nu = 0.5,kernel="rbf", gamma=0.8)
```

```python
model_lasso.fit(X,y)
#scores_lasso_pred = model_lasso.decision_function(X) 
y_lasso_pred = model_lasso.predict(X)
```

```python
y_lasso_pred
```

```python
y
```

Wektor błędów

```python
e=abs(y-y_lasso_pred)
e
```

10% outlierów - największe wartości z wektora błędów

```python
e2=e.argsort()[-44:][::-1]
```

```python
e2
```

```python
e[262]
```

```python
ground_truth = np.digitize(e,bins=[0,e[262]])
```

```python
ground_truth[ground_truth==2]=-1
```

```python
ground_truth
```

Modele SVM i CLF

```python
model_svm = svm.OneClassSVM(nu = 0.5,kernel="rbf", gamma=0.8)
```

```python
model_svm.fit(X)
scores_svn_pred = model_svm.decision_function(X) 
y_svm_pred = model_svm.predict(X)
```

```python
clf = LocalOutlierFactor(n_neighbors=30)
y_clf_pred = clf.fit_predict(X)
scores_clf_pred = clf.negative_outlier_factor_
```

```python
y_svm_pred
```

```python
y_clf_pred
```

Sprawdzenie pod kątem zbieżności wyników

```python
n_svm_errors = (y_svm_pred != ground_truth).sum()
```

```python
n_clf_errors = (y_clf_pred != ground_truth).sum()
```
